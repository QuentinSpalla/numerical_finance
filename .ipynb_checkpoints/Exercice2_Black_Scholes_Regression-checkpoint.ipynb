# The call and put options

We recall that a call is an European option that gives a payoff function equal to:

\begin{equation}
\left( S_T - K \right)^+
\end{equation}

Alternatively, a put is an European option that gives a payoff function equal to:

\begin{equation}
\left( K - S_T \right)^+
\end{equation}

Its parameters are the maturity $T$ and the strike $K$, and of course the considered underlying $S$.

We also recall the Call-Put parity (model free):

\begin{equation}
C_t - P_t = S_t - K D(t, T)
\end{equation}

where $D(t, T)$ is the price at time $t$ of the Zero-coupon bond i.e. $D(t, T) = \mathbb{E}_t^{\mathbb{Q}} \left( e^{-\int_t^T r_s \, \mathrm{d} s} \right)$

# The Black-Scholes model

In the Black-Scholes model, the underlying is defined by the following SDE:
\begin{equation}
\begin{cases}
\frac{\mathrm{d} S_t}{S_t} = \mu \, \mathrm{d} t + \sigma \, \mathrm{d} W_t \\
S_0 = x
\end{cases}
\end{equation}

whereas the riskless asset is defined by:
\begin{equation}
\begin{cases}
\mathrm{d} S_t^0 = r S_t^0 \, \mathrm{d} t \\
S_0^0 = 1
\end{cases}
\end{equation}

# Global imports

```python
import numpy as np
from scipy.stats import norm
```

# Closed formulas


We have, for $t \leq T$:
\begin{equation}
    C_t = S_t \mathcal{N} \left( d_+ \right) - K e^{-r (T - t)} \mathcal{N} \left( d_- \right)
\end{equation}
with: $d_{\pm} = \frac{1}{\sigma \sqrt{T - t}} \left[
\ln \left( \frac{S_t}{K} \right) + \left( r \pm \frac{\sigma^2}{2} \right) (T - t)
\right]$ where $\mathcal{N}$ is the standard Gaussian CDF.


```python
def bs_call_price(S=100., r=0.02, sigma=0.2, t=0., T=1., K=100.):
    ttm = T - t
    
    if ttm < 0:
        return 0.
    elif ttm == 0.:
        return np.maximum(S - K, 0.)

    vol = sigma * np.sqrt(ttm)

    d_minus = np.log(S / K) + (r - 0.5 * sigma**2) * ttm
    d_minus /= vol

    d_plus = d_minus + vol

    res = S * norm.cdf(d_plus)
    res -= K * np.exp(-r * ttm) * norm.cdf(d_minus)

    return res
```

```python
print(bs_call_price())
```

```python
def bs_put_price(S=100., r=0.02, sigma=0.2, t=0., T=1., K=100.):
    # Using call-put parity :)
    ttm = T - t
    
    if ttm < 0:
        return 0.
    elif ttm == 0.:
        return np.maximum(K - S, 0.)

    dsct_strike = K * np.exp(-r * ttm)

    cap_S = S * np.exp(r * t)

    call = bs_call_price(S, r, sigma, t, T, K)
    return call - cap_S + dsct_strike
```

```python
print(bs_put_price())
```

# Monte Carlo algorithms


The Black-Scholes price is given by:
\begin{equation}
C_t = \mathbb{E}_t^{\mathbb{Q}} \left( e^{-r (T -t) } \left( S_T - K \right)^+ \right)
\end{equation}
where $\mathbb{Q}$ denotes the (unique) risk neutral probability measure.


```python
def bs_call_price_mc(S=100., r=0.02, sigma=0.2,
                     t=0., T=1., K=100.,
                     nb_paths=int(1e4)):

    ttm = T - t
    discount_factor = np.exp(-r * ttm)

    sum_, sum2_ = 0., 0.
    for i in range(nb_paths):
        gauss = np.random.normal()
        log_ST = (r - 0.5 * sigma**2) * ttm + sigma * np.sqrt(ttm) * gauss

        payoff = discount_factor * np.maximum(S * np.exp(log_ST) - K, 0.)

        sum_ += payoff
        sum2_ += payoff**2

    mean = sum_ / nb_paths
    var = (sum2_ / nb_paths - mean**2) / (nb_paths - 1)

    return mean, var
```

```python
print(bs_call_price_mc())
```

```python
def bs_call_price_mc_opt(S=100., r=0.02, sigma=0.2,
                         t=0., T=1., K=100.,
                         nb_paths=int(1e4)):

    ttm = T - t
    discount_factor = np.exp(-r * ttm)

    gauss_mean = (r - 0.5 * sigma**2) * ttm
    gauss_stddev = sigma * np.sqrt(ttm)

    gauss_rvs = np.random.normal(
                    loc=gauss_mean,
                    scale=gauss_stddev,
                    size=nb_paths)

    payoffs = discount_factor * np.maximum(S * np.exp(gauss_rvs) - K, 0.)

    mean = np.mean(payoffs)
    var = payoffs.var() / (nb_paths - 1)

    return mean, var
```

```python
print(bs_call_price_mc_opt())
```

```python
%timeit bs_call_price_mc(nb_paths=int(1e5))
%timeit bs_call_price_mc_opt(nb_paths=int(1e5))
```

## Global algorithms

```python
def bs_mc_payoff(payoff_func, S=100., r=0.02, sigma=0.2,
                 t=0., T=1., nb_paths=int(1e4)):

    ttm = T - t
    discount_factor = np.exp(-r * ttm)

    gauss_mean = (r - 0.5 * sigma**2) * ttm
    gauss_stddev = sigma * np.sqrt(ttm)

    gauss_rvs = np.random.normal(
                    loc=gauss_mean,
                    scale=gauss_stddev,
                    size=nb_paths)

    vec_payoff = np.vectorize(payoff_func)
    payoffs = discount_factor * vec_payoff(S * np.exp(gauss_rvs))

    mean = np.mean(payoffs)
    var = payoffs.var() / (nb_paths - 1)

    return mean, var
```

```python
def call_payoff(s):
    return np.maximum(s - 100., 0.)

print(bs_mc_payoff(call_payoff))
```

```python
class BlackScholesMCPricer(object):
    def payoff(self, vec_s):
        raise NotImplementedError()

    def price(self, S=100., r=0.02, sigma=0.2, t=0., T=1., nb_paths=int(10e5)):
        ttm = T - t
        discount_factor = np.exp(-r * ttm)

        gauss_mean = (r - 0.5 * sigma**2) * ttm
        gauss_stddev = sigma * np.sqrt(ttm)

        gauss_rvs = np.random.normal(
                    loc=gauss_mean,
                    scale=gauss_stddev,
                    size=nb_paths)

        payoffs = discount_factor * self.payoff(S * np.exp(gauss_rvs))

        mean = np.mean(payoffs)
        var = payoffs.var() / (nb_paths - 1)

        return mean, var
```

```python
class BlackScholesMCCallPricer(BlackScholesMCPricer):
    def __init__(self, notional, strike):
        self.__not = notional
        self.__strk = strike
        self.__vect_payoff = np.vectorize(self.payoff_)
        
    def payoff_(self, s):
        return self.__not * np.maximum(s - self.__strk, 0.)
        
    def payoff(self, vec_s):
        return self.__vect_payoff(vec_s)
```

```python
mc_call_pricer = BlackScholesMCCallPricer(1., 100.)
print(mc_call_pricer.price())
```

# Black-Scholes processes

```python
def simulate_bs_process(S0=100., r=0.02, sigma=0.2, T=1., h=1./360):
    nb_sim = int(T / h) - 1
    
    if nb_sim <= 0:
        return [S0]
    
    gauss_mean = (r - 0.5 * sigma**2) * h
    gauss_stddev = sigma * np.sqrt(h)
    
    gauss_rvs = np.random.normal(
        loc=gauss_mean,
        scale=gauss_stddev,
        size=nb_sim)
    
    exp_gauss = np.exp(gauss_rvs)
    exp_gauss = np.insert(exp_gauss, 0, S0)

    return np.cumprod(exp_gauss)   
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

import seaborn as sns
sns.set_context('talk')
```

```python
nb_paths = 100

time_vec = np.linspace(0., 1., 360)
sim_paths = np.array([simulate_bs_process() for _ in range(nb_paths)])
```

```python
for i in range(nb_paths):
    plt.plot(time_vec, sim_paths[i])
    
plt.xlabel('Time (in years)')
plt.ylabel(r'Values of $S$')
plt.title('Samples of Geometric Brownian motion')
plt.show()
```


When $r = 0$ we have that $(S_t)_{t \geq 0}$ is a $\mathbb{Q}$ martingale so
\begin{equation}
\mathbb{E}^{\mathbb{Q}} \left( S_t \right) = S_0
\end{equation}
for all $t$.
Moreover:
\begin{equation}
\mathbb{V}ar^{\mathbb{Q}}(S_t) = S_0^2 \left( e^{\sigma^2 t} - 1 \right)
\end{equation}


Let's see the behaviour numerically:

```python
nb_paths = int(1e5)
```

```python
sim_paths_no_drifts = np.array([simulate_bs_process(r=0.) for _ in range(nb_paths)])

def bs_variance_th(S0, sigma, t):
    return S0**2 * (np.exp(sigma**2 * t) - 1)

plt.plot(time_vec, sim_paths_no_drifts.mean(axis=0), label='Simulated Mean')
plt.plot(time_vec, bs_variance_th(100., 0.2, time_vec), label='Theoretical Variance')
plt.plot(time_vec, sim_paths_no_drifts.var(axis=0), label='Simulated Variance')

plt.legend(loc='best')
plt.show()
```

```python
perc = 0.5

plt.plot(time_vec, sim_paths_no_drifts.mean(axis=0), label='Simulated Mean')
plt.plot(time_vec, np.percentile(sim_paths, 100 - perc, axis=0), label='%s %% quantile' % (100 - perc))
plt.plot(time_vec, np.percentile(sim_paths, perc, axis=0), label='%s %% quantile' % (perc))

plt.legend(loc='best')
plt.show()
```
